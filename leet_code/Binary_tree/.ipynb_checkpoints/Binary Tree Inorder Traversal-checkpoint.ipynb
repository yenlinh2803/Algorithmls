```python
# Recursion solution:

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        # if root==None:
        #     return None
        out_put = []
        if root:
            print("root:",root)
            out_put = self.inorderTraversal(root.left)
            out_put.append(root.val)
            out_put = out_put + self.inorderTraversal(root.right)
        return out_put
    
```

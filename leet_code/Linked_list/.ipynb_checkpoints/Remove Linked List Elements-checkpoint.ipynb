```python
# Linh code solution

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: ListNode, val: int) -> ListNode:
        if head == None :
            return None
        
        new_node =  ListNode(val= 123, next = head)
        head = new_node.next
        previous_node = new_node
        
        if previous_node.next.val == val and head.next== None:
            return None
        elif previous_node.next.val == val and head.next.val== val and head.next.next== None:
            return None
        else:
            while head!= None:
                if previous_node.next.val == val:
                    if previous_node.next.next is not None:
                        head.val = head.next.val
                        head.next = head.next.next 
                    else:
                        previous_node.next= None 
                        head = previous_node.next
                        return new_node.next
                else:
                    previous_node = head
                    head = head.next 
            return new_node.next
```

```python
# Online solution
class Solution:
    def removeElements(self, head: ListNode, val: int) -> ListNode:
        dummy = ListNode(next=head)
        prev, curr = dummy, head
        
        while curr:
            nxt = curr.next
        
            if curr.val == val:
                prev.next = nxt
            else:
                prev = curr
            curr = nxt
			
        return dummy.next
```

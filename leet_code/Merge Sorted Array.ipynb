```python
# Solution 2: Better 
# Test case 1: 
nums1 = [-1,0,0,3,3,3,0,0,0]
m =6
nums2= [1,2,2]
n=3
# MEMBER LEETCODE SOLUTION:
i= m-1
j= n-1
k= m+n-1
while i>-1 and j>-1:
    print("nums1 = ",nums1)
    print("nums2 = ",nums2)
    print("m= {} & n={}".format(m,n))
    print("i= {} & j={}".format(i,j))
    print("nums1[{}]= {} & nums2[{}]={}".format(i,nums1[i],j,nums2[j]))
    if nums2[j] >= nums1[i]:
        nums1[k] = nums2[j]
        k -= 1
        j -= 1
    else:
        nums1[k] = nums1[i]
        k -= 1
        i -= 1
    print("nums1 after = ",nums1)
    print("i={} & j={} & k={}".format(i,j,k))
    print("-"*30)
while j > -1:
    nums1[k] = nums2[j]
    k -= 1
    j -= 1
```
